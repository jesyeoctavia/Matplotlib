# Simple Scatter Plot

```python
x = [2, 4, 6, 8, 10, 11, 11.5, 11.7]
y = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5]
```

```python
plt.scatter(x, y, label='Data1', color='r')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Scatter Plot')

plt.legend()

plt.show()
```

# Multiple Scatter Plot

```python
x1 = [2, 4, 6, 8, 10, 11, 11.5, 11.7]
y1 = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5]

x2 = [8, 8.5, 9, 9.5, 10, 10.5, 11]
y2 = [3, 3.5, 3.7, 4, 4.5, 5, 5.2]
```

```python
plt.scatter(x1, y1, label='Data1', color='r')
plt.scatter(x2, y2, label='Data2', color='b')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Multiple Scatter Plot')

plt.legend()
plt.show()
```

# Pengaturan Marker

```python
x1 = [2, 4, 6, 8, 10, 11, 11.5, 11.7]
y1 = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5]

x2 = [8, 8.5, 9, 9.5, 10, 10.5, 11]
y2 = [3, 3.5, 3.7, 4, 4.5, 5, 5.2]
```

```python
plt.scatter(x1, y1,
           color='cyan',
           linewidths=1,
           marker='s',
           edgecolor='red',
           s=100)

plt.scatter(x2, y2,
           color='yellow',
           linewidths=1,
           marker='^',
           edgecolor='blue',
           s=100)

plt.show()
```

```python
plt.scatter(x1, y1,
           color='cyan',
           linewidths=1,
           marker='s',
           edgecolor='black', # atur warna garis tepi
           s=100)

plt.scatter(x2, y2,
           color='yellow',
           linewidths=1,
           marker='^',
           edgecolor='blue',
           s=100)

plt.show()
```

```python
plt.scatter(x1, y1,
           color='cyan',
           linewidths=1,
           marker='s',
           edgecolor='black', # atur warna garis tepi
           s=400)

plt.scatter(x2, y2,
           color='yellow',
           linewidths=1,
           marker='^',
           edgecolor='blue',
           s=100)

plt.show()
```

# Scatter Plot untuk data multi dimensi

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris['data'].T 
```

```python
plt.scatter(features[0], features[1],
            alpha=0.2,
            s=100*features[3],
            c=iris['target'],
            cmap='viridis')

plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
plt.title('Iris Dataset')

plt.colorbar()

plt.show()
```
